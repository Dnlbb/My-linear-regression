```python
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
import pandas as pd 
import numpy as np
df = pd.read_csv('/home/daniil/Документы/usa_mercedes_benz_prices.csv')
df.head(50)
```

```python
df['Mileage'] = df['Mileage'].astype(str)
df['Mileage'] = df['Mileage'].str.replace(' mi.', '').str.replace(',', '').astype(float)
df['Review Count'] = df['Review Count'].str.replace(',','').astype(float)
df = df.dropna(subset=['Rating'])
df = df[df['Price'] != 'Not Priced']
df['Price'] = df['Price'].str.replace('$', '').str.replace(',', '').astype(float)
df
```

```python
features = df[['Mileage', 'Rating', 'Review Count']]
target = df[['Price']]
features = np.array(features)
features
```

```python
target = np.array(target)
target = target.reshape(-1)
```

```python
def MseError_mat(X, w, y):
    y_pred = X @ w
    return  np.sum((y - y_pred) ** 2) / len(y_pred)
    
def gr_MseError_mat(X, w, y):
    y_pred = X @ w
    return 2 / len(X) * X.T @ (y_pred - y)
```

```python
weights = np.zeros(features.shape[1])
weights
```

```python
eps = 0.0001

learning_rate = 0.000000001

next_weights = weights
n = 1000
for i in range(n):
    cur_weights = next_weights
    next_weights = cur_weights - learning_rate * gr_MseError_mat(features, cur_weights, target)
    print(f"Iteration {i}")
    print(f"Current point {cur_weights}| Next point {next_weights}")
    print(f"MSE {MseError_mat(features, cur_weights, target)}")
    print("+++++++++++++++++++++++++++++++++++++++++++++++++")
    if np.linalg.norm(cur_weights - next_weights, ord = 2) <= eps:
        break
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
model = LinearRegression()
model.fit(features, target)
pred_target = model.predict(features)
mean_squared_error(target, pred_target)
```
